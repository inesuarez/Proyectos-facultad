```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#En el primer ejercicio hicimos que un vector random del mimso tama√±o que la matriz sea multiplicado por la matriz K veces y que cada uno de los autovalores aproximados para esos vectores los guarde en un vector a donde ai es el autovalor aproximado para el autovector de la iteracion i. 
def estado(A,k):
  n = A.shape[0]
  v = np.random.rand(n)
  a = []
  for i in range(k):
    v = (A @ v)/np.linalg.norm(v)
    a.append((np.transpose(v) @ A @ v)/(np.transpose(v) @ v))
  return a

```

```python
#ejemplo estado 
A = np.array([[1,0,0],[0,2,0],[0,0,3]])
print(estado(A,10))
print(estado(A,3))
print(estado(A,100))
```

```python
def generar(n):
  A = np.random.random((n,n))
  B2 = np.random.random((n,n))
  B = (np.transpose(B2)+ B2)/2
  C = B + 100* np.identity(B.shape[0])
  D = B + 1000* np.identity(B.shape[0])

  return A, B, C, D 
  #una funcion que nos genera la matriz A, una matriz random de nxn, la matriz B, una matriz random de nxn simetrica, 
  #la matriz C, que es igual a la matriz B pero sumandole 100 a cada elemento de la diagonal y la matriz D que es igual a la matriz B pero sumandole 1000 a cada elemento de la diagonal
```

```python
print(generar(100)) #muestra las 4 matrices de 100x100
```

```python
matrizA = generar(100)[0]
matrizB = generar(100)[1]
matrizC = generar(100)[2]
matrizD = generar(100)[3]
#fijamos cada una de las matrices usando la funcion generar
```

```python
print('esta es la matriz A de 100x100=\n',matrizA)
print('esta es la matriz B de 100x100=\n',matrizB)
print('esta es la matriz C de 100x100=\n',matrizC)
print('esta es la matriz D de 100x100=\n',matrizD)
#muestra cada una de las matrices
```

```python
def graficos(n,A,B,C,D):
  x = np.linspace(1, n, n) #con el linspace creamos un vector que ponga 100 numeros del 1 al 100 para representar las iteraciones
  A = plt.plot(x, estado(A,n)) 
  plt.xlabel('iteraciones') 
  plt.ylabel('autovalor aproximado') 
  plt.title('grafico autovalor matriz A') 
  plt.show()
  B = plt.plot(x, estado(B,n))
  plt.xlabel('iteraciones') 
  plt.ylabel('autovalor aproximado') 
  plt.title('grafico autovalor matriz B') 
  plt.show()
  C = plt.plot(x, estado(C,n)) 
  plt.xlabel('iteraciones') 
  plt.ylabel('autovalor aproximado') 
  plt.title('grafico autovalor matriz C') 
  plt.show()
  D = plt.plot(x, estado(D,n)) 
  plt.xlabel('iteraciones') 
  plt.ylabel('autovalor aproximado') 
  plt.title('grafico autovalor matriz D') 
  plt.show()
  return A, B, C, D
  #graficamos las iteraciones respecto a su autovalor aproximado para cada una de las matrices
```

```python
print(graficos(100,matrizA,matrizB,matrizC,matrizD)) #muestra los 4 graficos
```

Podemos concluir que mientras mas grande los numeros de la diagonal, mas grandes van a ser los autovalores y, por lo tanto, mas va a variar el grafico, ya que en los primeros 2 graficos converge muy rapido al mayor autovalor, y en los otros dos cuanto mas aumentamos los numeros de la diagonal mas lento fue convergiendo.
```python
def error(A,k):
  e = []
  max_av = max(np.linalg.eigvals(A))
  estados = estado(A,k)
  for i in range(k):
    p = abs(max_av - estados[i])
    e.append(p)
    #como hay que usar el mayor autovalor, no la aproximacion, usamos la funcion de autovalores de numpy en vez de la nuestra
  return e
#la funcion error nos da una lista con cada error del metodo de la potencia en funcion a cada iteracion
```

```python
#Ejemplo error
B = np.array([[1,0,0],[0,2,0],[0,0,3]])
print(error(B,3))
print(error(B,30))
print(error(B,100))
```

```python
def graficos_error(n,A,B,C,D):
  x = np.linspace(1, n, n)
  plt.plot(x, np.log(error(A,n))) #matrizA
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado') 
  plt.title('grafico error log matriz A') 
  plt.show()
  plt.plot(x, np.log(error(B,n))) #matrizB
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado') 
  plt.title('grafico error log matriz B') 
  plt.show()
  plt.plot(x, np.log(error(C,n))) #matrizC 
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado') 
  plt.title('grafico error log matriz C') 
  plt.show()
  plt.plot(x, np.log(error(D,n))) #matrizD  
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado') 
  plt.title('grafico error log matriz D') 
  plt.show()
  return 
#nos grafica el log del error aproximado en funcion a cada iteracion para cada matriz
```

```python
print(graficos_error(100,matrizA,matrizB,matrizC,matrizD))
#muestra los graficos 
```

```python
def grafico_recta(n,A,B,C,D):
  x=  np.linspace(1, n, n)
  wA = np.sort(np.absolute(np.linalg.eigvals(A))) #este vector nos ordena de menor a mayor los autovalores de la matriz, hay un w para cada una
  pA= 2*np.log(wA[-2]/wA[-1])*x #esta es la pendiente de la recta, hay una para cada matriz 
  oA= np.log(error(A,n)[1]) #esta la ordenada al origen, hay una para cada matriz
  eA= np.log(error(A,n)) #este es el grafico del error que hicimos anteriormente, hay uno para cada matriz
  rA = pA + oA #recta, hay una para cada matriz
  plt.plot(x , rA) #este plot me dibuja la recta
  plt.plot(x, eA) #este plot me dibuja el error
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado y recta') 
  plt.title('grafico error y recta matriz A') 
  plt.show() 
  wB = np.sort(np.absolute(np.linalg.eigvals(B)))
  pB= 2*np.log(wB[-2]/wB[-1])*x
  oB= np.log(error(B,n)[1])
  eB= np.log(error(B,n))
  rB = pB + oB
  plt.plot(x , rB)
  plt.plot(x, eB) 
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado y recta') 
  plt.title('grafico error y recta matriz B') 
  plt.show()
  wC = np.sort(np.absolute(np.linalg.eigvals(C)))
  pC= 2*np.log(wC[-2]/wC[-1])*x
  oC= np.log(error(C,n)[1])
  eC= np.log(error(C,n))
  rC = pC + oC
  plt.plot(x , rC)
  plt.plot(x, eC) 
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado y recta') 
  plt.title('grafico error y recta matriz C') 
  plt.show() 
  wD = np.sort(np.absolute(np.linalg.eigvals(D)))
  pD= 2*np.log(wD[-2]/wD[-1])*x
  oD= np.log(error(D,n)[1])
  eD= np.log(error(D,n))
  rD = pD + oD
  plt.plot(x , rD)
  plt.plot(x, eD) 
  plt.xlabel('iteraciones') 
  plt.ylabel('error aproximado y recta') 
  plt.title('grafico error y recta matriz D') 
  plt.show() 
  #en estos graficos juntamos el grafico del error de la funcion anterior con la recta tangente para cada una de las matrices 
```

```python
print(grafico_recta(100,matrizA,matrizB,matrizC,matrizD))
```

Podemos concluir que en el grafico de la matriz A el error aporximado converge rapido al error del ultimo autovalor aproximado, porque al converger rapido al mayor autovalor, rapidamente el error va a alcanzar su valor mas chico, ya que el error depende del autovalor. Ademas la pendiente es la que mas rapido se va para abajo porque los dos ultimos autovalores no son muy parecidos.

Podemos concluir que en el grafico de la matriz B, simetrica, el error aporximado converge rapido al error del ultimo autovalor aproximado ya que en el metodo de la potencia tambien se converge rapido al mayor autovalor pero antes de converger es muy parecido a la recta tangente, cosa que en la matriz A no pasaba. La pendiente aca tambien se va muy rapido para abajo.

En la matriz C, al aumentar los valores de la diagonal, aumentan los autovalores (como probamos en un ejercico de la practica3), el tiempo que le lleva al metodo de la potencia converger y, por lo tanto, el tiempo que le lleva al error converger.
